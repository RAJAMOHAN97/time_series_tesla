```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))
    if result[1]<=0.05:
          print("strong evidence againest the null hypothesis, reject the null hypothesis and its stationary data")
    else:
          print("week evidence againest the null hypothesis,time series has a unit root and its a non stationary data")
```

```python
adf_test(df["Open"])
```

```python
df["Open_fitst_diffwrwnce"]=df["Open"]-df["Open"].shift(1)
```

```python
df.tail()
```

```python
adf_test(df["Open_fitst_diffwrwnce"].dropna())
```

```python
df["Open_12_diffwrwnce"]=df["Open"]-df["Open"].shift(12)
```
